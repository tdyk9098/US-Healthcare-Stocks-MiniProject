```python
import pandas_datareader.data as web
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
import plotly
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode,plot,iplot
init_notebook_mode(connected = True)
import chart_studio.plotly as py
import scipy.stats as st
cf.go_offline()
%matplotlib inline
```

```python
unh = web.DataReader("UNH","yahoo",start = "2012-01-01", end = "2022-01-01")
wba = web.DataReader("WBA","yahoo",start = "2012-01-01", end = "2022-01-01")
```

```python
unh["Close"].iplot()
```

```python
# Correlation and P Val for when UNH took over in returns (2018), check MiniProject file for Rolling % Change
close = unh.reset_index()[unh.reset_index()["Date"]>"01/01/2018"].set_index("Date")["Close"]
dt = unh.reset_index()[unh.reset_index()["Date"]>"01/01/2018"]["Date"].view(int)
st.pearsonr(close,dt)
```
